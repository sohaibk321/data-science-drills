```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
%matplotlib inline
```

```python
sns.set(style='ticks')
df = pd.read_csv('avocado.csv')
df2 = pd.read_csv('consolidated_weather_V03.csv')
```

```python
# 1 continuous 1 categorical 6 ways
f, axes = plt.subplots(3, 2, figsize=(10,8))

#box plot: gives us a decent first look, but doesn't give exact details about the distribution
sns.boxplot(x='type', y='AveragePrice', data=df, ax=axes[0][0])

#violin plot used to estimate probability density function as well as the info from a regular boxplot
sns.violinplot(x='type', y='AveragePrice', data=df, ax=axes[0][1])

#strip plot shows us the distribution of data points, but doesn't give us a clear picture other than range
#one side note, there are organic avocados that are as cheap as regular avocados
sns.stripplot(x='type', y='AveragePrice', data=df, ax=axes[1][0])

#fourth plot
sns.boxenplot(x='type', y='AveragePrice', data=df, ax=axes[1][1])

sns.pointplot(x='type', y='AveragePrice', data=df, ax=axes[2][0])

sns.barplot(x='type', y='AveragePrice', data=df, ax=axes[2][1])
plt.show()
```

```python
#1 var 4 ways
f2, ax = plt.subplots(4,1, figsize=(20,20))
#shows us the relative distribution of when UFO sightings occur, doesn't tell us centrality or variance
sns.countplot(x='hour', data=df2, ax=ax[0])

#shows the spread and range but doesn't have some of the deeper measures such as variance or density of distribution
sns.boxplot(x='hour', data=df2, ax=ax[1])

#a good supplement to the boxplot to show us th density of the data
sns.violinplot(x='hour', data=df2, ax=ax[2])

#shows us a distribution of the points, although doesn;t give us a better picture about spread or centrality
sns.scatterplot(x='hour', y=df2.hour.value_counts(), data=df2, ax=ax[3])
plt.show()
```

```python
# two continuous
f3, ax2 = plt.subplots(3,1, figsize=(17,10))

#shows us a simple trend of how the price has changed but not much else
sns.lineplot(x='year', y='AveragePrice', hue='type', data=df, ax=ax2[0])

#shows us the distribution of price by type and year but any insight is blurred by the size of the distribution(too large)
sns.scatterplot(x='year', y='AveragePrice', hue='type', data=df, ax=ax2[1])

#shows us a good distribution of average price by year and gives us a spread too. Doesn't show variance that well
sns.boxplot(x='year', y='AveragePrice', hue='type', data=df, ax=ax2[2])
```
